```python
categorical_columns = df.select_dtypes(include=['object']).columns.drop('text')
categorical_columns
```

```python
for col in categorical_columns:
    if df[col].nunique()<11:
        one_hot = pd.get_dummies(df[col], prefix=col, drop_first=True)
        df = pd.concat((df.drop(col, axis=1), one_hot), axis=1)
        
    else:
        mean_target = df.groupby(col)['target'].mean()
        df[col] = df[col].map(mean_target)       
```

```python
def age_group(age):
    if age < 14:
        return 0
    elif age <= 19:
        return 1
    elif age <= 24:
        return 2
    elif age <= 32:
        return 3
    elif age <= 45:
        return 4
    else:
        return 5

    
df['age_group'] = df['age'].apply(lambda x: age_group(x))
df = df.drop('age', axis = 1)


df = df.rename(columns={'city': 'city_score'})
```

```python
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from collections import Counter
import string

# Создаем TF-IDF векторайзер с использованием топ 100 часто встречаемых слов
tfidf_vectorizer = TfidfVectorizer(max_features=1000)
tfidf_matrix = tfidf_vectorizer.fit_transform(df['text'])

# Создаем датафрейм с TF-IDF признаками
tfidf_df = pd.DataFrame(tfidf_matrix.toarray(), columns=tfidf_vectorizer.get_feature_names_out())

# Объединяем исходный датафрейм с TF-IDF признаками
df = pd.concat([df, tfidf_df], axis=1)
```

```python
df = df.drop('text', axis=1)
```

```python
X = df.drop("target", axis=1)
y = df["target"]


### Отделим валидацию от теста по времени!

X_test, X_train = X[X.timestamp >= '2021-12-11'], X[X.timestamp < '2021-12-11']

y_test, y_train = y[y.index.isin(X_test.index)], y[y.index.isin(X_train.index)]

### Уберем колонку с датой
X_test = X_test.drop('timestamp', axis=1)
X_train = X_train.drop('timestamp', axis=1)
X = X.drop('timestamp', axis=1)
```

```python
from catboost import CatBoostClassifier

catboost_model = CatBoostClassifier(learning_rate=0.02)

catboost_model.fit(X_train, y_train)

catboost_model.predict(X_train)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score

y_pred = catboost_model.predict(X_test)
y_pred_tr = catboost_model.predict(X_train)
y_pred_proba = catboost_model.predict_proba(X_test)[:, 1]

accuracy = accuracy_score(y_test, y_pred)
accuracy_train = accuracy_score(y_train, y_pred_tr)
precision = precision_score(y_test, y_pred)
recall = recall_score(y_test, y_pred)
f1 = f1_score(y_test, y_pred)

print(f"Accuracy: {accuracy:.4f}")
print(f"Accuracy of train data: {accuracy_train:.4f}")
print(f"Precision: {precision:.4f}")
print(f"Recall: {recall:.4f}")
print(f"f1-score: {f1:.4f}")
```

```python
# Получение предсказаний модели для тестовой выборки
predictions = catboost_model.predict(X_test)  # предсказания вероятностей классов

hit_rate_5 = 0
total_samples = len(y_test)

for true_label, pred_probs in zip(y_test, predictions):
    top_5_indices = pred_probs.argsort()[-5:][::-1]  # индексы топ-5 предсказаний
    if true_label in top_5_indices:
        hit_rate_5 += 1

hit_rate_5 /= total_samples

print("HitRate@5:", hit_rate_5)
```
